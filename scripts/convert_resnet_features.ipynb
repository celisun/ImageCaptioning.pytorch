```python
# convert resnet features scv to h5 files 
# input prepro before model model training
import h5py
import pandas as pd
import numpy as np

import os

csv_fc = 'features_resnet1000_train.csv' 
csv_att = 'features_resnet1000intermediate_train.csv'
dir_fc = ''
dir_att = ''
target_name_fc = 'data_fc.h5'  # save features after 1000 fc layer
target_name_att = 'data_att.h5' # save features after the last conv layer, 2048, 1, 1

# read resnet featues csv
tmp_fc = pd.read_csv(csv_fc, header=None).values   # shape 10000, 1001        
tmp_att = pd.read_csv(csv_att, header=None).values # shpae 10000, 2049 
i_fc = {}
i_att = {}
for i, e in enumerate(tmp_fc):
    i_fc[e[0]] = i
for i, e in enumerate(tmp_att):
    i_att[e[0]] = i
# # sort by image id
# tmp_fc = tmp_fc[np.argsort(tmp_fc[:, 0])]
# tmp_att = tmp_att[np.argsort(tmp_att[:, 0])]
```

```python
with h5py.File(os.path.join(dir_fc, 'data_fc.h5')) as file_fc,\
       h5py.File(os.path.join(dir_att, 'data_att.h5')) as file_att:        
        for n in range(len(tmp_fc)):
            ids = "{}.jpg".format(n)
            imgid = "images_train/{}.jpg".format(n)
            i, j = i_fc[imgid], i_att[imgid]
            assert(tmp_att[j][0] == tmp_fc[i][0])
            assert(tmp_att[j][0] == imgid)
            
            imgid = tmp_fc[i][0]
            print (imgid)
            d_set_fc = file_fc.create_dataset(ids, 
              (1000,), dtype="float")
            
            d_set_att = file_att.create_dataset(ids, 
              (1,1,2048), dtype="float")

            d_set_fc[...] = tmp_fc[i][1:].astype("float")  # 1000
            d_set_att[...] = tmp_att[j][1:].reshape((1,1,-1)).astype("float")  # 1 1 2048 
            if i % 1000 == 0:
                print('processing %d/%d (%.2f%% done)' % (i, 10000, i*100.0 / 10000))
        file_fc.close()
        file_att.close()
    
```
